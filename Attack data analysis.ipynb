## Analysis of attack victims

```python
#remove all ''s from list. make sure no ''s in the list
while '' in victims:
    victims.remove('')
```

2. count the occurrance of each victim

```python
victims_count = {}

for item in victims:
    if item in victims_count:
        victims_count[item] += 1
    else:
        victims_count[item] = 1
```

#### Second, I would like to count the types of attacks (espionage, sabotage, data destruction, defacement, denial of service) for each victim.

```python
#take attack types and victims from original data and make them into a table
type_table = pd.DataFrame(data,columns=['Victims','Type'])
```

1. count espionage attacks

```python
#in type_table, only keep the rows that is Espionage
espionage_table = type_table[type_table.Type == 'Espionage']

# in espionage_table, only keep the ones that have no missing value
espionage_table = espionage_table[espionage_table.Victims != '']
```

```python
#print the first 5 records in the table
espionage_table.head(5)
```

```python
#take Victims column and split based on ','
espionage_victims = []
for item in espionage_table['Victims']:
    if item != '':
        espionage_victims += item.split(',')
```

```python
#clean up the extra whitespaces around victim names
for i in range(len(espionage_victims)):
    espionage_victims[i] = espionage_victims[i].strip()
```

```python
#remove all ''s from list. make sure no ''s in the list
while '' in espionage_victims:
    espionage_victims.remove('')
```

```python
#count how many times each victims occurred
espionage_count = {}

for item in espionage_victims:
    if item in espionage_count:
        espionage_count[item] += 1
    else:
        espionage_count[item] = 1
```

```python
#make dictionary into table
espionage_count_table = pd.DataFrame(espionage_count.items(),columns = ['Espionage victims','count'])
espionage_count_table = espionage_count_table.sort_values(by=['count'], ascending=False, ignore_index= True)
```

```python
#export into excel
espionage_count_table.to_excel('result.xlsx', sheet_name = 'espionage', index = False)
```
